```python
from utils import RandomForest, get_columns, get_predict_clolumns
```

### 第一问

```python
# 使用随机森林进
RF = RandomForest(
    train_file_path = '附件一（训练集）.xlsx',
    predict_file_path = '附件二（测试集）.xlsx',
    train_start_column_id = 4,
    predict_start_column_id = 4
)
selected_columns1, selected_columns2, selected_columns3, selected_columns4 = get_columns('附件一（训练集）.xlsx', 4)
predict1_column, predict2_column, predict3_column, predict4_column = get_predict_clolumns('附件二（测试集）.xlsx')
```

```python
# 材料一
print(f"{predict1_column}")
# RF.train(selected_columns1)
y_pred1 = RF.predict(selected_columns1, predict1_column, "材料1")
y_pred1
```

```python
# 材料二
print(f"{predict2_column}")
# RF.train(selected_columns2)
y_pred2 = RF.predict(selected_columns2, predict2_column, "材料2")
y_pred2
```

```python
# 材料三
print(f"{predict3_column}")
# RF.train(selected_columns3)
y_pred3 = RF.predict(selected_columns3, predict3_column, "材料3")
y_pred3
```

```python
# 材料4
print(f"{predict4_column}")
# RF.train(selected_columns4)
y_pred4 = RF.predict(selected_columns4, predict4_column, "材料4")
y_pred4
```

```python
import numpy as np
# 将它们依次拼接
y_pred_combined = np.concatenate((y_pred1, y_pred2, y_pred3, y_pred4), axis=0)

# 结果的形状
print(y_pred_combined.shape)  
print(y_pred_combined)
```

```python
# 进行映射
output = np.zeros(y_pred_combined.shape[0], dtype=int)
output[np.all(y_pred_combined == [True, False, False], axis=1)] = 2  # 对应 [1,0,0] 三角
output[np.all(y_pred_combined == [False, True, False], axis=1)] = 3  # 对应 [0,1,0] 梯形
output[np.all(y_pred_combined == [False, False, True], axis=1)] = 1  # 对应 [0,0,1] 正弦

print(output)
```

```python
from collections import Counter

count = Counter(output)

# 打印结果
print(count)
```

```python
#三角波44                   1表示正弦波，2表示三角波，3表示梯形波
#正弦20
#梯形16
```

```python
# 写入附件4 第二列
import pandas as pd
df = pd.read_excel('附件四（Excel表）.xlsx')

# 将 output 列表转换为 DataFrame
output_df = pd.DataFrame(output, columns=['Values'])

# 将输出写入 DataFrame 的第二列
df.iloc[:len(output), 1] = output_df['Values']

# 将更新后的 DataFrame 保存回 Excel
df.to_excel('附件四（Excel表）.xlsx', index=False)
```
